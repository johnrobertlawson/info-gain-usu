```python
import numpy as np 

class BS:
    def __init__(self,o,f,dec_places=2):
        self.o = o 
        self.f = f
        self.I = np.round(sorted(set(self.f)), dec_places)
        
    def compute_rel(self):
        rel_score = 0  
        for yi in self.I:
            pyi = np.count_nonzero(self.f == yi)/len(self.f)
            zi = self.o[self.f == yi]
            zi_bar = zi[zi == 1] / len(zi)
            print(f"{yi=}")
            print(f"{pyi=}")
            print(f"{zi=}")
            print(f"{zi_bar=}")
            rel_score += pyi * (yi - zi_bar)**2
        return rel_score
    
    def compute_res(self):
        res_score = 0
        for yi in self.I:
            pyi = np.count_nonzero(self.f == yi)/len(self.f)
            z_bar = np.count_nonzero(self.o == 1)/len(self.o)
            zi = self.o[self.f == yi]
            zi_bar = np.count_nonzero(zi == 1) / len(zi)
            res_score += pyi * (zi_bar - z_bar)**2
        return res_score
    
    def compute_unc(self):
        z_bar = np.count_nonzero(self.o == 1)/len(self.o)
        return z_bar * (1-z_bar)
    
    def compute_bs(self):
        return np.mean((self.f - self.o)**2)

        
            
```

```python
def generate_time_series(seed=42,num_samples=10000):
    rng = np.random.default_rng(seed=seed)

    o = rng.choice([0, 1], size=num_samples, p=[0.8, 0.2])
    
    f = np.zeros(num_samples)
    
    for i in range(num_samples):
        if o[i] == 1:
            f[i] = rng.choice(np.arange(0.70, 0.995, 0.01))
        else:
            f[i] = rng.choice(np.arange(0.01, 0.40, 0.01))
    f = np.clip(f,0.005,0.995)
    return o, f

o, f = generate_time_series()
print(o)
print(f)
```
